```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',500)
```

```python
df=pd.read_csv("ai_jobs_market_2025_2026.csv")
```

```python
is_null = df.isnull().sum()
print("null values inside columns\n\n",is_null)
```

```python
column_rows = df.shape
print("Total Rows :",column_rows[0])
print("Total columns :",column_rows[1])
```

```python
duplicated_rows = df.duplicated().sum()
print("duplicated rows count :",duplicated_rows)
```

```python
unique_columns=df.nunique()
print("unique data counts of columns\n")
print(unique_columns)
```

```python
for col in df.select_dtypes(include='object'):
    print(f"\n{col}:")
    print(df[col].value_counts())
```
